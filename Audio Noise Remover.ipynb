```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import noisereduce as nr
import soundfile as sf
```

### Importing file from drive


```python
# prompt: import from drive

from google.colab import drive
drive.mount('/content/drive')
```

```python
filepath = '/content/drive/MyDrive/DSP_PROJECTS/Audio_Noise_Removal/'
```

```python
noisy_audio = filepath + 'noise_train_AirportAnnouncements_1.wav'
signal, sr = librosa.load(noisy_audio, sr=None)
```

## Visualize the Noisy Signal

```python
plt.figure(figsize=(12, 4))
librosa.display.waveshow(signal, sr=sr)
plt.title('Original Noisy Audio Waveform')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.show()
```

#### The visual above shows the time-domain waveform of the audio.


## Applying Spectral Gating Noise Reduction

```python
# The algorithm automatically estimates the noise profile
cleaned_signal = nr.reduce_noise(y=signal, sr=sr, prop_decrease=1.0)
```

## Visualize the Denoised Signal

```python
plt.figure(figsize=(12, 4))
librosa.display.waveshow(cleaned_signal, sr=sr)
plt.title('Denoised Audio Waveform (Spectral Gating)')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.show()
```

## Save the Denoised Audio

```python
output_file = filepath + 'cleaned_airport_spectral_gate.wav'
sf.write(output_file, cleaned_signal, sr)

print(f"Denoised audio saved as {output_file}")
```
